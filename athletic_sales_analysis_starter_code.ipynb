### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
invoice_date_total_sales_pivot = pd.pivot_table(mwaf_sold_retailer, values='total_sales', index=['invoice_date'], aggfunc='sum', fill_value=None).rename(columns={'total_sales':'Total Sales'})


# Optional: Rename the "total_sales" column to "Total Sales"


# Show the table.
invoice_date_total_sales_pivot.head()
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_bins = invoice_date_total_sales_pivot.resample('D').sum()
daily_bins
# Sort the resampled pivot table in ascending order on "Total Sales".
```

```python
sorted_daily_bins = daily_bins.sort_values(by='Total Sales', ascending=True)
sorted_daily_bins
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_bins = invoice_date_total_sales_pivot.resample('W').sum()
weekly_bins
# Sort the resampled pivot table in ascending order on "Total Sales".
```

```python
sorted_weekly_bins = weekly_bins.sort_values(by='Total Sales', ascending=True)
sorted_weekly_bins
```
